```python
for pack in pa1:
    #requests each package
    res = requests.get("https://pypi.org/pypi/{pack}/json".format(**locals()))
    if(str(res)=="<Response [404]>"):
        unavailable.append(pack)
        release_dates.append("")
    else:
        d = res.json()
        d.keys()

        #finds the first released version of the package
        i=2
        v_char = str(d['releases'])
        version = ""
        while v_char[i] != ':':
            version += v_char[i]
            i+=1

        #removes the last ' from the str
        version = version[0:len(version)-1]
        
        if(d['releases'][version] == []):
            unavailable.append(pack)
            release_dates.append("")
        else:
            #takes the date from the dictionary file
            #print(pack + " " + d['releases'][version][0]['upload_time'])
            release_dates.append((d['releases'][version][0]['upload_time'])[0:10])
```
